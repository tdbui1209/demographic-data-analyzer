```python
import pandas as pd
import numpy as np


df = pd.read_csv('adult.data.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['race'].value_counts()
```

```python
average_age_men = (df[df['sex'] == 'Male'])['age'].agg(np.mean)
average_age_men
```

```python
have_bachelors = df[df['education'] == 'Bachelors'].value_counts()

percentage_bachelors = len(have_bachelors) / len(df['education']) * 100
percentage_bachelors
```

```python
higher_education = df[df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])]

percentage_higher_education = len(higher_education) / len(df['education']) * 100
percentage_higher_education
```

```python
lower_education = df[~df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])]

percentage_lower_education = len(lower_education) / len(df['education']) * 100
percentage_lower_education
```

```python
df['salary'].unique()
```

```python
higher_education_rich = len(higher_education[higher_education['salary'] == '>50K']) / len(higher_education) * 100
higher_education_rich
```

```python
lower_education_rich = len(lower_education[lower_education['salary'] == '>50K']) / len(lower_education) * 100
lower_education_rich
```

```python
min_work_hours = df['hours-per-week'].min()
min_work_hours
```

```python
min_workers = df[df['hours-per-week'] == min_work_hours]
min_workers
```

```python
num_min_workers = len(min_workers)
rich_percentage = len(min_workers[min_workers['salary'] == '>50K']) / num_min_workers * 100
rich_percentage
```

```python
df['native-country'].unique()
```

```python
print(f"Percentage of unknown native country: {len(df[df['native-country'] == '?']) / len(df['native-country']) * 100}")
```

```python
x = pd.crosstab(df['native-country'], df['salary'], margins=True).apply(lambda r: r/r.sum() *2, axis=1)
x.style.background_gradient(cmap='autumn_r')
```

```python
print(f"highest percentage of people that earn >50K: {x['>50K'].max()}")
```

```python
print(f"country has the highest percentage of people that earn >50K: {x[x['>50K'] == x['>50K'].max()].index.values[0]}")
```

```python
rich_indian = df[(df['salary'] == '>50K') & (df['native-country'] == 'India')]
indian_occupation = (rich_indian['occupation'].value_counts())
indian_occupation
```

```python
top_IN_occupation = indian_occupation[indian_occupation == indian_occupation.max()].index[0]
top_IN_occupation
```
